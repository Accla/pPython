# Setting Up pPython Runtime Environment
<strong>This Jupyter notebook assumes that </strong>
<li>pPython package is already installed. </li>
<li>pPython configuration is already done. </li>


```python
# Import PythonMPI launch funciton
import pPython
from pRUN import *

# Disable HDF5 file locking (Lustre parallel filesystem on LLSC)
os.environ["HDF5_USE_FILE_LOCKING"]="FALSE"

# By default, local filesystem based messaging kernel is enabled.
# For interactive job, disable it.
os.environ['PPYTHON_LOCAL_FS'] = 'no'
```

# Writing pPython Program: Parameter Sweep

## Change the current working directory to "Param Sweep" tutorial

```python
import os
HOME = os.getenv('HOME')
os.chdir(HOME+os.sep+'ppython_tutorial'+os.sep+'Param_Sweep')
    
print(os.getcwd())
```

## Code section: global and local properties for pPython
<br>Extract the global and local properties for pPython process
<br><strong>Print "My process ID (Pid or Rank)"</strong>
<br><strong>Print "How many pPython processes are running (Np)"</strong>

### Code

```python
src_file_v1 = 'param_sweep_v1.py'
print('Read file, %s'%(src_file_v1))
print('-------------')
f = open(src_file_v1, "r")
print(f.read())
f.close()
```

### Run param_sweep_v1.py

```python
# Define number of MPI processes
n_proc = [1, 4, 1]

# Launch pPython
pRUN( src_file_v1, n_proc, 'grid' )
```

```python
# Check the PythonMPI subdirectory, which is automatically created
for item in sorted(os.listdir('PythonMPI')):
    print(item)
```

```python
#Check the standard output from the other processes
my_list = os.listdir('PythonMPI')
for entry in my_list:
    if re.search('p1-',entry):
        #Find the subdirectory for the first compute node 
        subdir_name = entry
        break
for i in range(3):
    fname = "PythonMPI"+os.sep+subdir_name+os.sep+"param_sweep_v1."+str(i+1)+".out"
    print('File: %s'%(fname))
    print('-------')
    fid = open(fname,"r")
    print(fid.read())
    fid.close()
```
